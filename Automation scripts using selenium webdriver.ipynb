```python
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.proxy import Proxy, ProxyType
import time
from bs4 import BeautifulSoup
import requests
import urllib.request as url
from random import choice
from selenium.webdriver.chrome.options import Options
```

```python
service = ChromeService(executable_path=ChromeDriverManager().install())
driver=webdriver.Chrome(service=service)
driver.get("https://www.yellowpages.com.au/")
driver.implicitly_wait(10)
time.sleep(3)
links=[]
categories=[]
services=driver.find_elements(by=By.XPATH , value="//div[@class='rsTmb icon-container']")
for service in services:
    
    link=service.find_element(by=By.CLASS_NAME,value="homepage-quicklinks").get_attribute("href")
    print(link)
    links.append(link)
    category=service.find_element(by=By.CLASS_NAME,value="icon-title").get_attribute("textContent")
    print(category)
    categories.append(category)
```

```python
import pandas as pd
m=0
df=pd.DataFrame([],columns=["Category","Names","Location","Contact","Website Links"])

for _link in links[0:2]:
    driver.get(_link)
    driver.implicitly_wait(10)
    time.sleep(3)
    
    Category=[]
    #names=[]
    #addresses=[]
    #phonenumbers=[]
    #ratings=[]
    #websitelinks=[]
    
    boxes=driver.find_elements(by=By.XPATH,value="//div[@class='Box__Div-sc-dws99b-0 dAyAhR']")
    
    #for box in boxes:
    #    print(box.text)
     #   name=box.find_element(by=By.XPATH, value="//h3[contains(@class,'MuiTypography-h3 MuiTypography-displayBlock')]").get_attribute("textContent")
     #   if(name!=""):
     #       names.append(name)
     #   else:
     #       names.append("no name")
    
    #print(names)
    
    Names=[]
    for i in range(1,1+len(boxes)):
        Category.append(categories[m])
        titles=driver.find_elements(By.XPATH, "//*[@id='root']/div/div[1]/div[2]/div/div[1]/div[2]/div[{}]/div[1]/div/div[1]/a/h3".format(i))
        if titles:
            Names.append(titles[0].text)
        else:
            Names.append('no name found')
    #print(Names)
    
    Phoneno=[]
    for i in range(1,1+len(boxes)):
        data=driver.find_elements(By.XPATH, "//*[@id='root']/div/div[1]/div[2]/div/div[1]/div[2]/div[{}]/div[2]/div/div/div/div[2]/div/div[1]/div[2]/button/span[1]".format(i))
        if data:
            Phoneno.append(data[0].text)
        else:
            Phoneno.append('no contacts found')
            
    #print(Phoneno) 
    
    Links=[]
    for i in range(1,1+len(boxes)):
        link=driver.find_elements(By.XPATH, "//*[@id='root']/div/div[1]/div[2]/div/div[1]/div[2]/div[{}]/div[2]/div/div/div/div[2]/div/div[2]/a".format(i))
        if link:
            Links.append(link[0].get_attribute("href"))
        else:
            Links.append('no website links')
    #print(Links) 
    
    Address=[]
    

    for i in range(1,1+len(boxes)):
        locate=driver.find_elements(By.XPATH, "//*[@id='root']/div/div[1]/div[2]/div/div[1]/div[2]/div[{}]/div[1]/div/div[1]/div/p".format(i))
        if locate:
            Address.append(locate[0].text)
        else:
            Address.append('no location found')
    #print(Address)
    print(len(Category),len(Names),len(Address),len(Phoneno),len(Links))
    data1={"Category":Category,"Names":Names,"Location":Address,"Contact":Phoneno,"Website Links":Links}
    tempdf=pd.DataFrame(data1)
    df=pd.concat([df,tempdf])
    m=m+1
    
```
