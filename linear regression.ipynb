```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

df = pd.read_csv('downloads/SaratogaHouses.csv')
x=df['livingArea'].values
y=df['price'].values

y
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
x_mean=np.mean(x_train)
y_mean=np.mean(y_train)
n=0
d=0
print(len(x_train))
print(len(y_train))
x_train
```

```python
for i in range(len(y_train)):
    n+=(y_train[i]-y_mean)*(x_train[i]-x_mean)
    d+=(x_train[i]-x_mean)*(x_train[i]-x_mean)
    
slope=n/d
c=y_mean-slope*x_mean

reg=LinearRegression()
x_train=x_train.reshape(-1,1)
y_train=y_train.reshape(-1,1)
reg.fit(x_train,y_train)
x_test=x_test.reshape(-1,1) #-1 means rows are unknown and numpy will adjust them as per requirement
resultx=reg.predict(x_test)
resultx
```

```python
y_pred=slope*x_test+c
```
