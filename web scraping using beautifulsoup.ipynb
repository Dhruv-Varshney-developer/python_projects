```python
team1=[]
team2=[]
winner=[]
#to get the text
m=[]


for j in y:
    z=j.find_all(class_="left")
    for a in z:
        m.append(a.get_text())

m        
```

```python
team1.append(m[::5])
team1
```

```python
team2.append(m[1::5])
team2
```

```python
winner.append(m[2::5])
winner
```

```python
finalteam1=team1[0]
finalteam2=team2[0]
finalwinner=winner[0]
data=[finalteam1,finalteam2,finalwinner]
import pandas as pd
df=pd.DataFrame(data,index=["Team 1","Team2","Winner"])
df1=df.T
df1
```

```python
count=df1["Winner"].value_counts()
count
```

```python
count.plot(kind='pie')
```

```python
type(count)
```

```python
count1=count.drop(index=["abandoned","no result","tied"])
count1
```

```python
count1.plot(kind='pie',autopct='%1.1f%%',label="Win %")
```

```python
ties=[1,0,0,1,0,0,0,0,0]
totalpoints=[15,14,14,13,10,6,6,6,4]
countdf=pd.DataFrame({'Teams':count1.index, 'Win count':count1.values, 'total win points': count1.values*2,'ties': ties, 'total points':totalpoints })
countdf
```

```python
import matplotlib.pyplot as plt
plt.bar(countdf['Teams'],countdf['total points'] , color ='maroon',
        width = 0.4) 
plt.xlabel("Teams")
plt.ylabel("Total points")
plt.title("Bar graph")
plt.rcParams["figure.figsize"] = (20,20)
plt.show()
```
