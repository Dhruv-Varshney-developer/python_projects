```python
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.proxy import Proxy, ProxyType
import time
```

```python
from bs4 import BeautifulSoup
import requests
import urllib.request as url
from random import choice
from selenium.webdriver.chrome.options import Options

path="https://free-proxy-list.net/"
page = requests.get(path)
soup = BeautifulSoup(page.content, 'html.parser')
x=soup.find(class_="table table-striped table-bordered")
y=x.find_all("td")
import pandas as pd
x=soup.find(class_="table table-striped table-bordered")
y=x.find_all("td")
ipaddress=[]
port=[]
https=[]
c=1 # c is used as element's address in list
for i in y:
    if(c%8==1):
        ipaddress.append(i.text)
    elif(c%8==2):
        port.append(i.text)
    elif(c%8==7):
        https.append(i.text)
    c=c+1
data={"IP Address": ipaddress, "Port": port , "HTTPS": https}
df=pd.DataFrame(data)
#https://IPaddress:portid- standard format of ip address
https=[]
http=[]
for index,row in df.iterrows(): #iterrows()- returns each row as a series and index number 
    
    if(row["HTTPS"]=="yes"):
        https.append(row["IP Address"]+ r":" + row["Port"])
    else:
        http.append(row["IP Address"]+ r":" + row["Port"])

random_proxy=choice(http)
print(random_proxy)
chrome_options = Options()
chrome_options.add_argument('--proxy-server=%s' % random_proxy)
#prox = Proxy()
#prox.proxy_type = ProxyType.MANUAL
#prox.http_proxy = random_proxy
#prox.socks_proxy = random_proxy
#prox.ssl_proxy = random_proxy

#capabilities = webdriver.DesiredCapabilities.CHROME
#prox.add_to_capabilities(capabilities)

#print(http)
```

```python


service = ChromeService(executable_path=ChromeDriverManager().install())
#driver = webdriver.Chrome(service=service,desired_capabilities=capabilities)
driver=webdriver.Chrome(service=service)
driver.get("https://www.yellowpages.com.au/")
driver.implicitly_wait(100)
time.sleep(5)
services=driver.find_elements(by=By.XPATH , value="//div[contains(@class, 'rsTmb icon-container')]")
#allservice=set()
#print(services)
allservice=[]
headingofservice=[]
nameinservice=[]
addressinservice=[]
phonenumbersinservice=[]
ratingsinservice=[]
websitelinkinservice=[]
#for service in services:
    
    
for service in services[1:]:
    names=[]
    addresses=[]
    phonenumbers=[]
    ratings=[]
    websitelinks=[]
    #print("x")
    allservice.append(service.text)
    service.click()
    driver.implicitly_wait(8)
    
    Names=[]
    for i in range(2,37):
    titles=driver.find_elements(By.XPATH, "//*[@id='root']/div/div[1]/div[2]/div/div[1]/div[2]/div[{}]/div[1]/div/div[1]/a/h3".format(i))
    if titles:
        Names.append(titles[0].text)
    else:
        Names.append('no location found')
   
    
    y=driver.find_elements(by=By.XPATH, value="//a[contains(@class,'MuiTypography-root MuiLink-root MuiLink-underlineNone MuiTypography-colorPrimary')]")
    for name in y:
        #print(name.text)
        if(name.text!=""):
            names.append(name.text)
        else:
            names.append("no name")
        
      
        
    box=driver.find_element(by=By.XPATH, value="//div[contains(@class,'Box__Div-sc-dws99b-0 fWSsVI')]")   
    
    z=box.find_elements(by=By.XPATH, value="//div[contains(@class,'Box__Div-sc-dws99b-0 bKFqNV')]")
    for address in z:
        #print(address.text)
        if(address.text!=""):
            addresses.append(address.text)
        else:
            addresses.append("no address")
        
       
        
       # phonenumberbox=box.find_element(by=By.XPATH, value="//button[contains(@class,'MuiButtonBase-root MuiButton-root MuiButton-text MuiButton-textPrimary MuiButton-fullWidth')]")
        
      #  phonenumber=phonenumberbox.find_element(by=By.XPATH,value="//span[contains(@class,'MuiButton-label')]")

            
    a=driver.find_elements(by=By.XPATH, value="//button[contains(@class,'MuiButtonBase-root MuiButton-root MuiButton-text MuiButton-textPrimary MuiButton-fullWidth')]")
    for phonenumber in a:
        #print(phonenumber.text)
        if(phonenumber.text!=""):
            phonenumbers.append(phonenumber.text)
        else:
            phonenumbers.append("no phonenumber found")
        
    
    b=driver.find_elements(by=By.XPATH, value="div[contains(@class,'MuiTypography-root jss283 MuiTypography-body2 MuiTypography-colorTextSecondary')]")
    for rating in b:
        #print(rating.text)
        if(rating.text!=""):
            ratings.append(rating.text)
        else:
            ratings.append("no ratings found")
            
    c=driver.find_elements(by=By.XPATH, value="//a[contains(@class,'MuiButtonBase-root MuiButton-root MuiButton-text MuiButton-textSecondary MuiButton-fullWidth')]")
    for websitelink in c:
        d=websitelink.get_attribute("href")
        #print(d)
        if(d!=""):
            websitelinks.append(d)
        else:
            websitelinks.append("no website link found")
        
        
        
    print(names)
    print('\n')
    print(addresses)
    print('\n')
    print(phonenumbers)
    print('\n')
    print(websitelinks)
    print('\n')
    nameinservice.append(names)
    addressinservice.append(addresses)
    phonenumbersinservice.append(phonenumbers)
    websitelinkinservice.append(websitelinks)
    print(nameinservice)
    print('\n')
    print(addressinservice)
    print('\n')
    print(phonenumbersinservice)
    print('\n')
    print(websitelinkinservice)
    print('\n')
    driver.quit()
    driver.back()
    driver.implicitly_wait(4)

#driver.find_element(by=By.CLASS_NAME, value="rsThumbsArrow rsThumbsArrowRight").click()
time.sleep(10)
#services2=driver.find_elements(by=By.XPATH , value="//div[contains(@class, 'rsNavItem rsThumb')]")
#for service in services2:
#    allservice.add(service.text)
print(allservice)
print("\n")
print(headingofservice)
print("\n")
print(nameinservice)
print("\n")
print(addressinservice)
print("\n")
print(phonenumbersinservice)
print("\n")
print(websitelinkinservice)
print("\n")
```
