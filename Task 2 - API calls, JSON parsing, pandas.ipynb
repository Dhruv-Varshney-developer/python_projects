```python
markers=dataset["markers"]
_address=[]
_lat=[]
_lon=[]
_title=[]
_zip=[]
_city=[]
_state=[]
additionaldata=[]
_phonenumber=[]
```

```python
import re
def getkeys1():
    for i in markers:
        address=i["address"]
        lat=i["lat"]
        lon=i["lng"]
        title=i["title"]
        customdata=i["custom_field_data"]
        
        _address.append(address)
        _lat.append(lat)
        _lon.append(lon)
        _title.append(title)
        additionaldata.append(customdata)
        
        zip=re.search(r"\d{5}|\d{5}-\d{4}",address)
        if zip is not None:
            zipfinal=zip.group()
            _zip.append(zipfinal)
        #city=re.search(r"",address).group()
        #_city.append(city)
        state=re.search(r"\B(NC)|(IN)|(AR)|(WV)|(MD)|(TN)|(KY)|(VA)|(OH)|(SC)",address)
        if state is not None:
            statefinal=state.group()
            _state.append(statefinal)
        
        for j in customdata:
            values=j["value"]
            phonenumbers=re.search(r"\d{3}-\d{3}-\d{4}| ^1-\d{3}-\d{3}-\d{4}",values)
            if phonenumbers is not None:
                phonenumbersfinal=phonenumbers.group()
                _phonenumber.append(phonenumbersfinal)
```
